# GPU 

```python
import cudf    
import cuml
import warnings
warnings.filterwarnings('ignore')
```

```python
train = cudf.read_csv('train.csv')

train
```

```python
from sklearn.model_selection import train_test_split
```

```python
from cuml.ensemble import RandomForestClassifier as RF_GPU
```

```python
X = train.drop(columns = ['Transported','PassengerID_1', 'First Name', 'Last Name', 'Cabin_1'])
y = train['Transported']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 612)
```

```python
rfModel_GPU = RF_GPU(n_estimators = 1000)

%timeit rfModel_GPU.fit(X_train, y_train)

y_pred_gpu = rfModel_GPU.predict(X = X_test)
```

# CPU

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier as RF_CPU
from sklearn.model_selection import train_test_split
```

```python
train_cpu = pd.read_csv('train.csv')

train_cpu
```

```python
X = train_cpu.drop(columns = ['Transported','PassengerID_1', 'First Name', 'Last Name', 'Cabin_1'])
y = train_cpu['Transported']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 612)
```

```python
rfModel_CPU = RF_CPU(n_estimators = 1000, n_jobs = -1)

%timeit rfModel_CPU.fit(X_train, y_train)

y_pred_cpu = rfModel_CPU.predict(X = X_test)
```
